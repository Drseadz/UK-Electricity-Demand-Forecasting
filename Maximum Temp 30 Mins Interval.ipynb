**Maximum Temperature Cleaning**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

**Year 2020**

```python
df_2020 = pd.read_csv('metoffice.gov.uk_hadobs_hadcet_data_cet_max_2020.txt', delimiter='\s+', skiprows=3)
df_2020 = df_2020.set_index('Date')
df_2020.head()
```

```python
df_2020.info()
```

```python
df_2020.tail()
```

```python
new_jan_values = [random.randint(1, 10) for _ in range(len(df_2020))]
df_2020['New_Jan'] = new_jan_values
```

```python
df_2020.loc[[30, 31], ['Feb']] = np.nan
df_2020.loc[[31], ['Apr', 'Jun', 'Sept', 'Nov']] = np.nan
```

```python
df1_2020 = pd.concat([df_2020['Jan'], df_2020['Feb'], df_2020['Mar'], df_2020['Apr'], df_2020['May'], df_2020['Jun'], 
                      df_2020['Jul'], df_2020['Aug'], df_2020['Sept'],df_2020['Oct'], 
                      df_2020['Nov'], df_2020['Dec'], df_2020['New_Jan']])
df2_2020 = pd.DataFrame(df1_2020, columns=['MAX_TEMP'])
df2_2020 =df2_2020.reset_index ()
df2_2020 = df2_2020.dropna()
#df2_2020 = df2_2020.drop(['DATE'], axis = 1)
df2_2020.head()
```

```python
from datetime import datetime
df2_2020['Date'] = pd.to_datetime(df2_2020['Date'], format='%d-%b-%Y', utc= True, infer_datetime_format=True)
date_range = pd.date_range(start='01-01-2020', end='31-01-2021', freq='D')
df2_2020['Date'] = df2_2020['Date'].dt.strftime('%d-%m-%Y')
df2_2020.set_index(date_range, inplace=True)
df2_2020=df2_2020.drop(['Date'], axis = 1)
df2_2020.head()
```

```python
df2_2020 = df2_2020.rename_axis('DATE')
df2_2020.head()
```

```python
df3_2020 = df2_2020.asfreq('30T').interpolate(method='linear').round(2)
df3_2020['HALFHOUR_PERIOD'] = ((df3_2020.index.hour * 60 + df3_2020.index.minute) / 30 + 1).astype(int)
df3_2020 = df3_2020.drop(df3_2020.tail(1441).index)
df3_2020.tail()
```

```python
df3_2020.info()
```

```python
df3_2020.to_csv('MaxTemp2020.csv')
```
