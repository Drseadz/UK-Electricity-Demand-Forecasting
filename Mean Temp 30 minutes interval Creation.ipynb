```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import random
```

**YEAR 2020**

```python
df = pd.read_csv('metoffice.gov.uk_hadobs_hadcet_data_cet_mean_2020.txt', delimiter='\s+', skiprows=3)
df= df.set_index('Date')
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
df.tail()
```

```python
new_jan_values = [random.randint(1, 10) for _ in range(len(df))]
df['New_Jan'] = new_jan_values
```

```python
df.loc[[30, 31], ['Feb']] = np.nan
df.loc[[31], ['Apr', 'Jun', 'Sept', 'Nov']] = np.nan
#df = df.replace(np.nan, '')
```

```python
df.tail()
```

```python
df1 = pd.concat([df['Jan'], df['Feb'], df['Mar'], df['Apr'], df['May'], df['Jun'], df['Jul'], df['Aug'], df['Sept'],
                df['Oct'], df['Nov'], df['Dec'],df['New_Jan']])
df2 = pd.DataFrame(df1, columns=['MEAN_TEMP'])
df2 =df2.reset_index ()
df2 = df2.dropna()
df2.head()
```

```python
from datetime import datetime
```

```python
df2['Date'] = pd.to_datetime(df2['Date'], format='%d-%b-%Y', utc= True, infer_datetime_format=True)
date_range = pd.date_range(start='01-01-2020', end='31-01-2021', freq='D')
df2['Date'] = df2['Date'].dt.strftime('%d-%m-%Y')
df2.set_index(date_range, inplace=True)
df2=df2.drop(['Date'], axis = 1)
df2.head(20)
```

```python
df2 = df2.rename_axis('DATE')
df2.head()
```

```python
df3 = df2.asfreq('30T').interpolate(method='linear').round(2)
df3['HALFHOUR_PERIOD'] = ((df3.index.hour * 60 + df3.index.minute) / 30 + 1).astype(int)
df3 = df3.drop(df3.tail(1441).index)
df3.tail()
```

```python
df3.info()
```

```python
df3.to_csv('MeanTemp2020.csv')
```
